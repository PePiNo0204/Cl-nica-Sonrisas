```python
#Funcion que salude
def saludar(nombre):
    print("Hola",nombre)
saludar("Galoc")
```

```python
#funcion doble numero
def doble(numero):
    return numero*2
print(doble(21))
```

```python
#Funcion suma de 3 numeros
def suma(a,b,c):
    return a+b+c
print(suma(3,4,21))
```

```python
#Funcion area de un regtangulo
def area_rectangulo(base,altura):
    return base*altura
print(area_rectangulo(3,6))
```

```python
#Funcion mayusculas
def mayusculas(palabra):
    return palabra.upper()
print(mayusculas("siuuuu"))
```

```python
#Funcion numero positivo
def positivo(numero):
    if numero > 0:
        return True
    else:
        return False
print(positivo(-5))
```

```python
#Funcion mayor de dos numeros
def mayor(a,b):
    if a > b:
        return a
    else:
        return b
print(mayor(45,23))
```

```python
# Funcion contar letras
def contar_letras(palabra):
    contador = 0
    for letra in palabra:
        contador = contador + 1
    return contador
print(contar_letras("Dinosaurio"))
```

```python
#Funcion promedio
def promedio(lista):
    suma = 0
    for numero in lista:
        suma = suma + numero
    return suma/len(lista)
print(promedio([12,17,19,14,15]))
```

```python
# funcion suma de lista
def suma_lista(lista):
    suma = 0
    for numero in lista:
        suma = suma + numero
    return suma
print(suma_lista([3,2,8,1,10]))
```

```python
#Funcion area de un circulo
def area_circulo(radio):
    pi = 3.1415
    return pi*radio**2
print(area_circulo(3))
```
